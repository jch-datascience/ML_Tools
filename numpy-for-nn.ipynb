**Ejercicios de trigonometría solo a modo de ejemplo**

```python
# utilizar np.linsace para crear un arreglo de 100 números entre 0 y 2π
from numpy import pi

ejercicio = np.linspace(0, 2*pi, 100)
ejercicio
```

```python
# calcular seno y coseno del arreglo anterior
seno = np.sin(ejercicio)
coseno = np.cos(ejercicio)
```

```python
# graficar un círculos con seno y coseno
plt.plot(seno, coseno)
plt.show()
```

**Operaciones con arreglos**

```python
# Crear dos arreglos
a = np.linspace(20, 30, 6)
b = np.linspace(30, 35, 6)

print(a, b)
```

```python
# Sumar los arreglos a y b
a + b
```

```python
# concatenar los arreglos 
concat = np.concatenate((a, b))
concat
```

```python
# ordenar los valores
concat.sort()
concat
```

**Arreglos Aleatorios**

```python
# generar 1000 valores random
rg = np.random.default_rng(2) # generador de números aleatorios con semilla 2
random = rg.random(1000) 
```

```python
# graficar valores
plt.hist(random, bins = 100)
plt.show()
```

```python
# generar 100000 valores con una distribución normal
normal_dist = rg.normal(10, 5, 100000) #media = 10, std = 5
```

```python
normal_dist
```

```python
# graficamos distribución normal
plt.hist(normal_dist, bins = 1000)
plt.show()
```

```python
# generar 2000 números aleatorios enteros entre 0 y 19
integers = rg.integers(20, size = 2000)
integers
```

```python
plt.hist(integers)
plt.show()
```

```python
# elegir 10 números aleatorios SIN REPETIR entre 0 y 25
rg.choice(26, size = 10, replace = False)
```
